# Layout Deep Dive Notebook (All Layouts)

This notebook explains every major layout surface in the current prototype, including frontend and admin workflows.


```python
from pathlib import Path
try:
    from IPython.display import SVG, Markdown, display
except Exception:
    def SVG(content):
        return content

    def Markdown(content):
        return content

    def display(content):
        print(content)

NOTEBOOK_DIR_CANDIDATES = [
    Path.cwd(),
    Path.cwd() / "docs" / "client-handoff" / "notebooks",
    Path("docs/client-handoff/notebooks"),
]
NOTEBOOK_DIR = next(
    (
        candidate.resolve()
        for candidate in NOTEBOOK_DIR_CANDIDATES
        if candidate.exists() and (candidate / "../assets").resolve().exists()
    ),
    Path.cwd().resolve(),
)


def resolve_notebook_path(path):
    target = Path(path)
    if target.is_absolute():
        return target

    return (NOTEBOOK_DIR / target).resolve()


def notebook_markdown_path(path):
    resolved = resolve_notebook_path(path)
    cwd = Path.cwd().resolve()
    try:
        return resolved.relative_to(cwd).as_posix()
    except ValueError:
        return resolved.as_posix()

def show_img(path, label='Image'):
    resolved = resolve_notebook_path(path)
    if resolved.exists():
        display(Markdown(f"![{label}]({notebook_markdown_path(path)})"))
    else:
        display(Markdown(f"**Missing {label}**\n\nExpected path: `{path}`\nResolved path: `{resolved}`"))

def md_list(items):
    return "\n".join([f"- {i}" for i in items])

def section_block(name, purpose, users, inputs, formulas, actions, good, risk, value, screenshot):
    txt = f"""### {name}

**Purpose**  
{purpose}

**Primary Users**  
{users}

**Inputs / Dependencies**  
{md_list(inputs)}

**Logic / Formula Notes**  
{md_list(formulas)}

**Operator Actions**  
{md_list(actions)}

**Healthy Signal**  
{good}

**Risk Signal**  
{risk}

**Client Value Story**  
{value}

**Screenshot Target**  
`{screenshot}`
"""
    display(Markdown(txt))
```
